```python
import pandas as pd

# 1. 데이터 로드
df_kaggle = pd.read_csv("/content/sample_data/survey.csv")        # Kaggle 버전
df_mendeley = pd.read_csv("/content/sample_data/data.csv")    # Mendeley 통합 버전

# 2. 컬럼명 정리 (예: 소문자화, 언더스코어로 변환 등)
df_kaggle.columns = df_kaggle.columns.str.strip().str.lower().str.replace(" ", "_")
df_mendeley.columns = df_mendeley.columns.str.strip().str.lower().str.replace(" ", "_")

# 3. 공통키 확인 및 병합할 키 선택
# 예시: country, age, gender 등이 공통일 수 있음
print("공통 컬럼:", set(df_kaggle.columns).intersection(df_mendeley.columns))

# 병합 키 예시
key_cols = ["country", "age", "gender"]

# 4. 병합 (inner / outer / left 등 전략 선택)
df_merged = pd.merge(
    df_kaggle,
    df_mendeley,
    on=key_cols,
    how="inner",    # 또는 "outer", "left" 등
    suffixes=("_kaggle","_mendeley")
)

# 5. 병합 후 처리
# 누락치 확인
print(df_merged.isna().sum())

# 필요시 누락치 처리 예:
df_merged = df_merged.dropna(subset=key_cols)   # 키가 없는 행 제거
df_merged.fillna({'some_column': 'Unknown'}, inplace=True)

# 6. 컬럼 중복 이름 정리
# 예: 'treatment_kaggle' 과 'treatment_mendeley' 둘 중 하나 선택
# df_merged['treatment'] = df_merged['treatment_mendeley'].combine_first(df_merged['treatment_kaggle']) # 'treatment_mendeley' 컬럼이 존재하지 않아 오류 발생. df_mendeley에 'treatment' 컬럼이 없음.
# df_merged.drop(columns=['treatment_kaggle','treatment_mendeley'], inplace=True)

# 'treatment' 컬럼은 df_kaggle에서 가져온 것으로 이미 df_merged에 존재합니다.
# df_mendeley에는 'treatment' 컬럼이 없으므로, 중복 처리 로직이 필요하지 않습니다.

# 7. 병합된 데이터 저장
df_merged.to_csv("merged_mental_health_survey.csv", index=False)
```
